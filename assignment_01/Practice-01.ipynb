### Duck Typing
Often you may not care about the type of an object but rather only whether it has
certain methods or behavior

```
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False
```

```
isiterable('a string')
```

```
isiterable([1, 2, 3])
```

#### for loops

```
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
```

```
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
```

```
for i in range(4):
    for j in range(4):
        if j > i:
            break
        print((i, j))
```

```
x = 256
total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
```
