#### Prompts a user to enter 10 integers

##### (Detects if user makes a mistake)

```
user_inputs=[]
for i in range(1, 11):
    while True:
        try:
            user_input=int(input("Please enter an integer"))
            break
        except ValueError:
            print("Sorry you must enter a number!")
            continue
    print(f"You have entered #{i}:", user_input)
    user_inputs.append(user_input)
print(user_inputs)
```

#### Minimum Value

```
min(user_inputs)
```

#### Maximum Value

```
max(user_inputs)
```

#### Range

```
range = (max(user_inputs)-min(user_inputs))
print(range)
```

```
import statistics
```

#### Mean

```
statistics.mean(user_inputs)
```

#### Variance

```
statistics.variance(user_inputs)
```

#### Standard Deviation

```
statistics.stdev(user_inputs)
```
