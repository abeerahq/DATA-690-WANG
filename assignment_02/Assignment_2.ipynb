#### Prompts a user to enter 10 integers

##### (Detects if user makes a mistake)

```
user_inputs=[] # Create an empty list.
for i in range(1, 11): # Specify a range.
    while True:
        try:
            user_input=int(input("Please enter an integer")) # Numeric integer is returned.
            break
        except ValueError: # Value Error allows user to retry if string is entered.
            print("Sorry you must enter a number!")
            continue
    print(f"You have entered #{i}:", user_input)
    user_inputs.append(user_input) 
    # All values entered are now put into the empty list created in the first step.
print(user_inputs) #Displays all 10 inputs.
```

#### Minimum Value

```
min(user_inputs)
```

#### Maximum Value

```
max(user_inputs)
```

#### Range

```
range = (max(user_inputs)-min(user_inputs))
print(range)
```

```
import statistics
```

#### Mean

```
statistics.mean(user_inputs)
```

#### Variance

```
statistics.variance(user_inputs)
```

#### Standard Deviation

```
statistics.stdev(user_inputs)
```
